```python
# Simply train a khiops classifier to distinguish train from test 
# Univariate importance (the level) of the features give "the drift" 
```

```python
import os
import pandas as pd
import numpy as np
import zipfile

from khiops import core as kh
from khiops.sklearn import KhiopsClassifier
```

```python
#First unzip the data
zip_path = 'UNSW_NB15_training-set.zip'
extract_path = './'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
zip_path = 'UNSW_NB15_testing-set.zip'
extract_path = './'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```

```python
# Load the data file into a Pandas DataFrame
X_train = pd.read_csv("UNSW_NB15_training-set.csv", sep=",")
X_test = pd.read_csv("UNSW_NB15_testing-set.csv", sep=",")
```

```python
X_train['IsDriftOrNot'] = 0
X_test['IsDriftOrNot'] = 1
```

```python
X_new=pd.concat([X_train, X_test], axis=0)
X = X_new.drop("IsDriftOrNot", axis=1)
y = X_new["IsDriftOrNot"]
```

```python
model = KhiopsClassifier(n_trees=0)
```

```python
model.fit(X,y)
```

```python
my_list = list(X)
df2 = pd.DataFrame({'nom_var': my_list})
df2['level'] = None
i=0
for variable in my_list:
    imp=model.model_report_.preparation_report.get_variable_statistics(variable).level
    df2.at[i,'level']=imp
    i=i+1
    #print(imp)
    #print(df2)
    df_trie = df2.sort_values(by='level',ascending=False)
print(df_trie)
```
